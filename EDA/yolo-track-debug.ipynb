```python
import ultralytics

# Load a pretrained YOLO model
model = ultralytics.YOLO("ckpt/best_100.pt")
model.metrics = ["mAP50", "mAP50-95", "precision"]
```

```python
model.train(
    project="live-scooters-content",
    data="datasets/live-scooters.yaml",
    device="0",
    single_cls=True,
    augment=True,
    shear=25,
    degrees=25,
    epochs=5,
    batch=-1,
    imgsz=640,
    plots=True,
    visualize=True,
    save_period=10,
    cos_lr=True,
    dfl=1.5,
    pose=0.0,
    label_smoothing=0.1,
    translate=0.3,
    mosaic=0.9,
    dropout=0.4,
    seed=0,
)
```

```python
# Validate the model
metrics = model.val(data="datasets/surveillance-cam-scooters.yaml")

# Access specific metrics
print(f"mAP50-95: {metrics.box.map}")
print(f"mAP50: {metrics.box.map50}")
print(f"mAP75: {metrics.box.map75}")
print(f"Precision (P): {metrics.box.mp}")
print(f"Recall (R): {metrics.box.mr}")
```
