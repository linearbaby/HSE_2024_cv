```python
from ultralytics import YOLO

# Load a pretrained YOLO model
model = YOLO('best_100.pt')  # You can use different YOLOv5 variants
```

```python
model.train(
    project="result/",
    data='datasets/kickscooter_dataset/data.yaml',
    device="cuda:0",
    augment=True,
    shear=50,
    degrees=20,
    epochs=100,
    batch=100,
    imgsz=420,
)
```
